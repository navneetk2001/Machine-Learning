**Question 1 (Simple Linear Regression) :-**

Q1) Implement the Simple Linear Regression for the data given below and predict the house price for a house with plot size 3000, 2000 and 1500 feet:-

# Method : Least Square Error Fit (LSE)- Step by Step

```python
# step 1 : Loading the Dataset
df1 = pd.DataFrame()
df1["S.No"] =  np.ones(10)
df1["Plot size(sq.ft)"] = [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]
df1["Price($1000s)"] = [245,312,279,308,199,219,405,324,319,255]
df1.info()
```

```python
# step 2 : Removing noise
import missingno as ms
ms.bar(df1)
```

```python
# step 2: Checking Redundancy among features
import seaborn as sns
sns.heatmap(df1.corr(),annot=True)
```

```python
# step 3 : splitting dataset in x and y and converting 1 d array to 2d array using reshape function
X1 = df1.iloc[:,0:2]
X1 = np.array(X1)

Y1 = df1.iloc[:,2]
Y1 = np.array(Y1)
Y1 = Y1.reshape(-1,1)

print(X1)
print(Y1)
```

```python
# step 5 : Finding Regression Coefficients
A1 = X1.T.dot(X1)
B1 = np.linalg.inv(A1)
C1 = B1.dot(X1.T)
beta1 = C1.dot(Y1)

print(beta1)
```

```python
# step 6: New testing Data to predict the house price for a house with given plot size
P1 = np.array([[1,3000],[1,2000],[1,1500]])
print(P1)
```

```python
#step 6 : Predicting values on test set
Q1 = P1.dot(beta1)
print(Q1)
```
